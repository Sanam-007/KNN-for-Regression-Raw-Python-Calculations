```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/data set/prl/weight-height.csv')
df.head()
```

```python
X = df[['Gender', 'Height']]
y = df['Weight']
```

```python
# Convert categorical variable 'Gender' to numerical using one-hot encoding
X = pd.get_dummies(X, columns=['Gender'], drop_first=True)

# Train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train) #training
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse_test_lr = mean_squared_error(y_test, y_test_pred_lr)

mse_test_lr
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Create the KNeighborsRegressor object
knn = KNeighborsRegressor(n_neighbors=5)

# Fit the model to the training data
knn.fit(X_train, y_train)

# Predict the target values for the test data
y_test_pred_knn = knn.predict(X_test)

# Calculate the MSE for the KNN Regressor
mse_test_knn = mean_squared_error(y_test, y_test_pred_knn)
y_test_pred_knn = knn.predict(X_test)
# Calculate MSE for KNN Regressor
mse_test_knn = mean_squared_error(y_test, y_test_pred_knn)
mse_test_knn
```
