```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
data = {'age': [21, 20, 22, 22, 23, 21, 25, 30, 31],
        'income': [60, 55, 60, 61, 65, 62, 65, 70, 68]}

df = pd.DataFrame(data)
df
```

```python
# Split the dataset into features (X) and target variable (y)
X = df[['age']]
y = df['income']
# KNN algorithm
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X, y)
# Predict income for X=22 using KNN
X_new = pd.DataFrame({'age': [22]})
knn_prediction = knn.predict(X_new)
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X, y)

# Predict income for X=22 using Linear Regression
lr_prediction = lr.predict(X_new)
```

```python
# Add the predictions to the dataframe
df['knn_prediction'] = knn.predict(X)
df['lr_prediction'] = lr.predict(X)
```

```python
# Display predictions
print(f'KNN Prediction for X=22: {knn_prediction[0]}')
print(f'Linear Regression Prediction for X=22: {lr_prediction[0]}')
```

```python
# Save the dataframe to an Excel file
df.to_excel('predictions_comparison.xlsx', index=False)
df
```
